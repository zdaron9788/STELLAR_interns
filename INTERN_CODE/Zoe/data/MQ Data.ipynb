```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path1 = open("Table_1_Periodic.txt", 'r')
path2 = open("Table_2_Non_Periodic.txt", 'r')
```

```python
path1_header = path1.readlines(10)
path1_header
```

```python
table1 = pd.read_csv("Table_1_Periodic.txt")
table1
```

```python
table2 = pd.read_csv("Table_2_Non_Periodic.txt")
table2
```

```python
plt.scatter(table1['Teff'], table1['Prot'], s = .3, c = 'r', label = "Table 1")
plt.scatter(table2['Teff'], table2['Prot'], s = .3, c = 'b', label = "Table 2")
plt.xlim(6500, 3500)
plt.title("Teff/Prot Graph")
```

```python
table1 = table1.dropna()
table2 = table2.dropna()
len(table1), len(table2) #compares lengths
```

```python
np.arange(0, 10, 1)
```

```python
valid = np.repeat(1, len(table1))
invalid = np.repeat(0, len(table2))
classes = np.append(valid, invalid)
```

```python
tables = pd.concat([table1, table2], axis = 0, sort = True)
tables = tables.drop(['Flag', 'Rper'], axis = 1)
```

```python
tables
```

```python
tables['Class'] = classes
tables
```

```python
plt.scatter(tables['Teff'], tables['Prot'], s = .3, c = tables['Class'], vmin = 0, vmax = 1)
plt.xlim(6500, 3500)
plt.title("Teff/Prot Graph")
plt.colorbar().set_label('Class', fontsize = 20)
plt.figure(figsize = (14, 8))
```

```python
tables = tables.drop(['Class'], axis = 1)
tables
```

```python
X = tables
y = classes

from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
print(model)
fit_model = model.fit(X, y)
print(fit_model)
```
